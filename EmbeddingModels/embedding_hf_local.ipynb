```python
from langchain_huggingface import HuggingFaceEmbeddings, HuggingFacePipeline
```

```python
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
query = "Shebaz is the prime minister of Pakistan"
result = embedding.embed_query(text=query)
print(result)
```

```python
print(str(result))
```

```python
document = [
    "Today weather is sunny",
    "Finally I start to learn LLMs",
    "Good to see you"
]
```

```python
document_result= embedding.embed_documents(texts=document)
print(document_result)
```

```python
print(str(document_result))
```
